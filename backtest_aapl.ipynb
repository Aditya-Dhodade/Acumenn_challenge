```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
data_of_aapl = yf.Ticker('AAPL')
data = data_of_aapl.history(period='1d', start='2000-1-1', end='2023-03-25')
```

```python
data
```

```python
data["Typical price"] = (data['High'] + data['Close'] + data['Low'])/3
```

```python
data['avg gain'] = 0
data['avg loss'] = 0
for i in range(1,5844):
    if (data['Typical price'][i] - data['Typical price'][i-1]) < 0:
        data['avg loss'][i] = abs(data['Typical price'][i] - data['Typical price'][i-1])
        data['avg gain'][i] = 0
    else:
        data['avg gain'][i] = abs(data['Typical price'][i] - data['Typical price'][i-1])
        data['avg loss'][i] = 0
```

```python
data['avgg'] = 0
data['avgl'] = 0

for i in range(14,5844):
    sumg = 0
    suml = 0
    for j in range(i-14,i):
        suml += data['avg loss'][j]
        sumg += data['avg gain'][i]
    data['avgg'][i] = sumg/14
    data['avgl'][i] = suml/14
```

```python
data['rs'] = 0
for i in range(14,5844):
    data['rs'][i] = data['avgg'][i]/data['avgl'][i]
```

```python
data['rsi'] = 0
for i in range(14,5844):
    data['rsi'][i] = 100 - (100 / (1 + data['rs'][i]))
```

```python
data['ema50'] = 0
data['ema200'] = 0
for i in range(50,5844):
    data['ema50'][i] = ((data['Close'][i] - data['ema50'][i-1]) * 2 / (51)) + data['ema50'][i-1]
for i in range(200,5844):    
    data['ema200'][i] = ((data['Close'][i] - data['ema200'][i-1]) * 2 / (201)) + data['ema200'][i-1]
    
```

```python
data['ema13'] = 0
data['ema26'] = 0
for i in range(13,5844):
    data['ema13'][i] = ((data['Close'][i] - data['ema13'][i-1]) * 2 / (14)) + data['ema13'][i-1]
for i in range(26,5844):    
    data['ema26'][i] = ((data['Close'][i] - data['ema26'][i-1]) * 2 / (27)) + data['ema26'][i-1]
```

```python
ema_26 = data['Close'].ewm(span=26, adjust=False).mean()
ema_12 = data['Close'].ewm(span=12, adjust=False).mean()
macd_line = ema_12 - ema_26
data['MACD'] = macd_line
```

```python
data = data.dropna()
data.info()
```

```python
data['result'] = 0
for i in range(200,5842):    
    buy = 0
    sell = 0
    if data['rsi'][i]>=30 and data['rsi'][i]<=60: buy += 1
    if data['ema50'][i] > data['ema200'][i]: buy += 1
    if data['ema13'][i] > data['ema26'][i]: buy += 1
    if data['MACD'][i] > 0: buy += 1
    
    if data['rsi'][i]>60: sell += 1
    if data['ema50'][i] < data['ema200'][i]: sell += 1
    if data['ema13'][i] < data['ema26'][i]: sell += 1
    if data['MACD'][i] < 0: sell += 1
        
    if buy >= 3:
        data['result'][i] = 1
    elif sell >= 3:
        data['result'][i] = 2
    else:
        data['result'][i] = 0
        
        
#     data['ema200'][i] = ((data['close'][i] - data['ema200'][i-1]) * 2 / (201)) + data['ema200'][i-1]
```
